```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
car_evaluation = fetch_ucirepo(id=19) 
  
# data (as pandas dataframes) 
X = car_evaluation.data.features 
y = car_evaluation.data.targets 
  
# metadata 
print(car_evaluation.metadata) 
  
# variable information 
print(car_evaluation.variables) 
```

```python
import pandas as pd
```

```python
df = pd.concat([X,y], axis=1)
df
```

```python
df["safety"].unique()
```

## H0 = Cars rated as "vgood"([class]) dont have higer safety score than cars with "good"([class]) rate 
## H1 = Cars rated as "vgood"([class]) have higer safety score than cars with "good"([class]) rate 

```python
vgood_class = df[df["class"] == "vgood"]
vgood_class["safety"].unique()
```

```python
saftey_scores = {
    "low": 1,
    "med": 2,
    "high": 3
}

vgood_class["safety"] = vgood_class["safety"].map(saftey_scores)

vgood_class.head()
```

```python
vgood_class.shape
```

```python
good_class = df[df["class"] == "good"]
good_class["safety"].unique()
```

```python
good_class["safety"] = good_class["safety"].map(saftey_scores)
good_class[:65].shape
```

```python
good_class["safety"]
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind( vgood_class["safety"], good_class["safety"])
```

## Conclusion:

T indicates, that the diffarence is high likely not just a noise. Besides, p value indicates, that the possibility of this outcome is so high that we can reject H0 (>0.05). Therefore, we can accept H1.
