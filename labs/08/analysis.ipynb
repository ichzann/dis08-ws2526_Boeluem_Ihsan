```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
car_evaluation = fetch_ucirepo(id=19) 
  
# data (as pandas dataframes) 
X = car_evaluation.data.features 
y = car_evaluation.data.targets 
  
# metadata 
print(car_evaluation.metadata) 
  
# variable information 
print(car_evaluation.variables) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
df = pd.concat([X, y], axis=1)
df
```

```python
count_by_classes = y["class"].value_counts()
```

```python
count_by_classes
```

```python
count_by_classes.index
```

```python
matplotlib.pyplot.bar(
    count_by_classes.index,
    count_by_classes
)
```

```python
df["class"].value_counts().plot(kind = "bar")
```

```python
df["buying"].value_counts()
```

```python
order = ["low", "med", "high", "vhigh"]
for i in df["class"].unique():
    cars_by_class = df[df[f"class"] == f"{i}"]
    counts = cars_by_class["buying"].value_counts().reindex(order).fillna(0)
    x_achse = counts.index
    plt.figure(figsize=(6, 4))
    plt.title(f"price by accountibility. Class = {i}")
    plt.ylabel("Number of cars")
    plt.xlabel("Buying Price")
    plt.bar(x_achse, counts, color = "teal")

```
