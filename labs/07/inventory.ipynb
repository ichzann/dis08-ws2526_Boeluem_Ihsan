# Comma Code

```python
def list_retrun(list):
    list_in_string = ""
    for i in list:
        if i != list[-1]:
            list_in_string += f"{i}, "
        else:
            list_in_string += f"{i}"
    return print(list_in_string)

        
    

list = ['apples', 'bananas', 'tofu', 'cats']
list_retrun(list)
```

# Coin Flip Streaks

```python
import random

def coin_flipper():

    H_T_List = [random.randint(0,1) for _ in range(100)]
    

    stats = {
        "current_head_streak": 0,
        "total_head_streaks": 0,
        "current_tail_streak": 0,
        "total_tail_streaks": 0
    }
   
    for i in H_T_List:
        

        if i == 0:
            stats["current_head_streak"] += 1   
            stats["current_tail_streak"] = 0    
            
            if stats["current_head_streak"] == 6:
                stats["total_head_streaks"] += 1
                stats["current_head_streak"] = 0 
        

        elif i == 1:
            stats["current_tail_streak"] += 1  
            stats["current_head_streak"] = 0    
            
            if stats["current_tail_streak"] == 6:
                stats["total_tail_streaks"] += 1
                stats["current_tail_streak"] = 0 

    return print(f"{stats['total_tail_streaks']} x 6 streak tail (1)\n{stats['total_head_streaks']} x 6 head streak (0)")

coin_flipper()
```

# Fantasy Game Inventory

Write a function named displayInventory() that would take any possible “inventory” and display it like the following:

Inventory:
12 arrow
42 gold coin
1 rope
6 torch
1 dagger
Total number of items: 62

```python
def inventory_sum(inventory):
    print("Inventory:")
    for amounth, keys in inventory.items():
        print(f"{amounth} {keys}")
    print(f"Total number of items: {sum(inventory.values())}")

inventory = {
    'rope': 1, 
    'torch': 6, 
    'gold coin': 42, 
    'dagger': 1, 
    'arrow': 12
    }

inventory_sum(inventory)
```

# List to Dictionary Function for Fantasy Game

```python
def addToInventory(inventory, loot):
    for i in loot:
        if i in inventory:
            inventory[i] += 1
        else:
            inventory[i] = 1


    return inventory

player_inventory = {
    'rope': 1, 
    'torch': 6, 
    'gold coin': 42, 
    'dagger': 1, 
    'arrow': 12
    }
dragonLoot = ['gold coin', 'dagger', 'gold coin', 'gold coin', 'ruby', "ruby"]
inv = addToInventory(player_inventory, dragonLoot)

inventory_sum(inv)
```
