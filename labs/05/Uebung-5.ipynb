```python
import requests
from bs4 import BeautifulSoup
import csv


teams_dic = []


def scraper():
    for page in range(1, 25):
        link = f"https://www.scrapethissite.com/pages/forms/?page_num={page}"
        response = requests.get(link)
        soup = BeautifulSoup(response.text, 'html.parser')

        all_wins = soup.find_all(class_="wins")
        all_years = soup.find_all(class_="year")
        all_teams = soup.find_all(class_="name")
        for year, win, name in zip(all_years, all_wins, all_teams):
            teams_dic.append({
                "year": year.text.strip(),
                "win": win.text.strip(),
                "name": name.text.strip()
            })
        
    return teams_dic
    
scraper()

csv_file = ".../hokey_teams.csv"

with open(csv_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["year", "wins", "name"])

    for team_data in teams_dic:
        writer.writerow([
            team_data["year"], 
            team_data["win"], 
            team_data["name"]
        ])
```
